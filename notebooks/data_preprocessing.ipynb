## Delays and cancellations per day of the week

### Selecting columns

```python
day_of_week = delays_cancellations[['fl_date','crs_dep_time','dep_time','dep_delay','cancelled']]
day_of_week.head()
```

```python
day_of_week['fl_date'] = pd.to_datetime(day_of_week['fl_date'], errors='coerce')
day_of_week.dtypes
```

A column with the day of the week is added

```python
day_of_week.insert(1, "fl_day_week", day_of_week['fl_date'].dt.day_name())
day_of_week.head()
```

```python
day_of_week['dep_delay'] = day_of_week['dep_delay'].fillna(0)
```

```python
day_of_week
```

Insert a column with value 1 when the flight is delayed and 0 when it's on time or advanced

```python
day_of_week.insert(5, "delay", np.where(day_of_week['dep_delay'] > 0, 1, 0))
day_of_week.head()
```

```python
day_of_week['delay'].value_counts()
```

```python
day_of_week['cancelled'].value_counts()
```

Insert a column with value 'On time' if the flight hasn't been delayed nor cancelled, 'Cancellation' when it has been cancelled, and 'Delay' in the rest

```python
day_of_week.insert(7, "delay_cancelled", np.where((day_of_week['delay'] == 0) & (day_of_week['cancelled'] == 0), 'On time', 
                                                  np.where(day_of_week['cancelled'] == 1,'Cancellation','Delay')))
day_of_week
```

New dataframe with frequencies. Merge of 4 dataframes grouped by the day of the week, and then the frequency columns are created

```python
correct_days_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
total_count_per_day = day_of_week.groupby(['fl_day_week']).agg({'fl_date':'count'}).rename(columns={"fl_date":"total_count"})
total_count_per_day = total_count_per_day.reindex(correct_days_order)

total_cancelled_per_day = day_of_week[day_of_week['delay_cancelled']=='Cancellation'].groupby(['fl_day_week']).agg({'fl_date':'count'}).rename(columns={"fl_date":"total_cancellations"}).reindex(correct_days_order)
total_delayed_per_day = day_of_week[day_of_week['delay_cancelled']=='Delay'].groupby(['fl_day_week']).agg({'fl_date':'count'}).rename(columns={"fl_date":"total_delays"}).reindex(correct_days_order)
total_on_time_per_day = day_of_week[day_of_week['delay_cancelled']=='On time'].groupby(['fl_day_week']).agg({'fl_date':'count'}).rename(columns={"fl_date":"total_on_time"}).reindex(correct_days_order)


freq_per_day = total_delayed_per_day.merge(total_cancelled_per_day,on='fl_day_week').merge(total_count_per_day,on='fl_day_week')
freq_per_day = freq_per_day.merge(total_on_time_per_day,on='fl_day_week')

freq_per_day['freq_delayed_day'] = freq_per_day['total_delays']/freq_per_day['total_count']
freq_per_day['freq_cancelled_day'] = freq_per_day['total_cancellations']/freq_per_day['total_count']
freq_per_day['freq_on_time'] = freq_per_day['total_on_time']/freq_per_day['total_count']
freq_per_day
```

```python
freq_per_day.reset_index(inplace=True)
```

```python
freq_per_day
```

### Save dataframes as csv

```python
day_of_week.to_csv(config["data"]["plot1"], index=False)
freq_per_day.to_csv(config["data"]["plot7"], index=False)
```

## Chance of diversion per destination airport

### Selecting columns

```python
diversion = delays_cancellations[['fl_date','dest','dest_airport','diverted']]
diversion.head()
```

```python
diversion['diverted'] = diversion['diverted'].astype(int)
diversion.head()
```

```python
diversion['diverted'].value_counts()
```

```python
total_count_per_dest = diversion.groupby(['dest']).agg({'fl_date':'count'}).rename(columns={"fl_date":"total_count"})

total_diverted_per_dest = diversion[diversion['diverted']==1.0].groupby(['dest']).agg({'fl_date':'count'}).rename(columns={"fl_date":"total_diverted"})
total_not_diverted_per_dest = diversion[diversion['diverted']==0.0].groupby(['dest']).agg({'fl_date':'count'}).rename(columns={"fl_date":"total_not_diverted"})

freq_per_dest = total_diverted_per_dest.merge(total_not_diverted_per_dest,on='dest').merge(total_count_per_dest,on='dest')

freq_per_dest['freq_diverted'] = freq_per_dest['total_diverted']/freq_per_dest['total_count']
freq_per_dest['freq_not_diverted'] = freq_per_dest['total_not_diverted']/freq_per_dest['total_count']
freq_per_dest
```

```python
freq_per_dest.reset_index(inplace=True)
```

### Save dataframe as csv

```python
diversion.to_csv(config["data"]["plot5"], index=False)
freq_per_dest.to_csv(config["data"]["plot11"], index=False)
```
